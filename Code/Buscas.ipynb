```python
class no :
    
    def __init__(self,valor,esquerda = None,direita = None):
        self.valor = valor
        self.esquerda = esquerda
        self.direita = direita
        
    def _getValor(self):
        return self.valor
    def _setValor(self,valor):
        self.valor = valor
        
    def _getAnterior(self):
        return self.esquerda.valor
    def _getProximo(self):
        if self.direita is not None:
            if self.direita.valor is not None:
                return self.direita.valor
        return None
    
    def _setAnterior(self,Novaesquerda):
        self.esquerda = Novaesquerda        
    def _setProximo(self,NovaDireita):
        self.direita = NovaDireita
        
    
```

```python
class lista :
    
    def __init__(self,head = None,lenght = 0):
        self.head = head
        self.lenght = 0
    
    def _getHead(self):
        return self.head
    
    def _getLenght(self):
        return self.lenght
    
    def _empty(self):
        return self.lenght == 0
    
    def _showList(self):
        
        if self._empty():
            
            return "Lista vazia"
        else:
            
            print("Tamnho da lista: ",self.lenght)
            
            #iterador = no(self.head.valor,self.head.esquerda,self.head.direita)
            iterador = self.head
            while iterador != None:
                print(iterador.valor)
                iterador = iterador.direita
            
                
    
    def _insertHead(self,novo):
        
        if self._empty():
            self.head = no(novo.valor,novo.esquerda,novo.direita)
            self.lenght += 1
            return "Inserido no inicio"
        else:
            novo._setProximo(self.head)
            #novo.direita = self.head
            self.head = novo
            self.lenght += 1
            return "inserido no inicio"
        
    def _insertMidle(self,novo,indice):
        
        if self._empty():
            print("Lista vazia ---> Inserindo no inicio")
            self._insertHead(novo)
        
        else :
            
            if indice > self.lenght or indice < 0:
                
                return "Index Error"
            
            else:
            
                iterador = self.head
                cont = 1
                
                while (cont < indice) and (iterador._getValor() != None):
                    iterador = iterador.direita
                    cont += 1
                
                
                novo._setProximo(iterador.direita)
                iterador._setProximo(novo)
                self.lenght += 1
                return "Inserido no meio"
            
    def _insertTail(self,novo):
        
        if self._empty():
            print("Lista vazia ---> Inserindo no inicio")
            self._insertHead(novo)
            
        else:
            
            iterador = no(self.head.valor,self.head.esquerda,self.head.direita)
            
            while iterador.direita != None:
                iterador._setProximo(iterador.direita)
                
            iterador._setProximo(novo)
            self.lenght += 1
            return "Inserido no fim"
            
            
```
