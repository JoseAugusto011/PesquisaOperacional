```python
class no :
    
    def __init__(self,valor,esquerda = None,direita = None):
        self.valor = valor
        self.esquerda = esquerda
        self.direita = direita
        
    def _getValor(self):
        return self.valor
    def _setValor(self,valor):
        self.valor = valor
        
    def _getAnterior(self):
        return self.esquerda.valor
    def _getProximo(self):
        if self.direita is not None:
            if self.direita.valor is not None:
                return self.direita.valor
        return None
    
    def _setAnterior(self,Novaesquerda):
        self.esquerda = Novaesquerda        
    def _setProximo(self,NovaDireita):
        self.direita = NovaDireita
        
    
```

```python
class lista :
    
    def __init__(self,head = None,lenght = 0):
        self.head = head
        self.lenght = 0
    
    def _getHead(self):
        return self.head
    
    def _getLenght(self):
        return self.lenght
    
    def _empty(self):
        return self.lenght == 0
    
    def _showList(self):
        
        if self._empty():
            
            return "Lista vazia"
        else:
            
            print("Tamnho da lista: ",self.lenght)
            
            #iterador = no(self.head.valor,self.head.esquerda,self.head.direita)
            iterador = self.head
            while iterador != None:
                print(iterador.valor)
                iterador = iterador.direita
            
                
    
    def _insertHead(self,novo):
        
        if self._empty():
            self.head = no(novo.valor,novo.esquerda,novo.direita)
            self.lenght += 1
            return "Inserido no inicio"
        else:
            novo._setProximo(self.head)
            #novo.direita = self.head
            self.head = novo
            self.lenght += 1
            return "inserido no inicio"
        
    def _insertMidle(self,novo,indice):
        
        if self._empty():
            print("Lista vazia ---> Inserindo no inicio")
            self._insertHead(novo)
        
        else :
            
            if indice > self.lenght or indice < 0:
                
                return "Index Error"
            
            else:
            
                iterador = self.head
                cont = 1
                
                while (cont < indice) and (iterador._getValor() != None):
                    iterador = iterador.direita
                    cont += 1
                
                
                novo._setProximo(iterador.direita)
                iterador._setProximo(novo)
                self.lenght += 1
                return "Inserido no meio"
            
    def _insertTail(self,novo):
        
        if self._empty():
            print("Lista vazia ---> Inserindo no inicio")
            self._insertHead(novo)
            
        else:
            
            iterador = self.head
            
            while iterador.direita != None:
                iterador = iterador.direita
                
            iterador._setProximo(novo)
            self.lenght += 1
            return "Inserido no fim"
        
        
    def _removeHead(self):
        
        if self._empty():
            return "Lista vazia"
        if self.lenght == 1:
            
            no = self.head
            self.head = None
            self.lenght -= 1
            
            return no

        else:
            
            no = self.head            
            self.head = self.head.direita
            no._setProximo(None)
            self.lenght -= 1
            
            return no
        
    def _removeTail(self):
        
        if self._empty():
            return "Lista vazia"
        
        if self.lenght == 1:
            
            return self._removeHead()
        
        else:
            
            iterador = self.head
            while iterador.direita.direita != None:
                iterador = iterador.direita
            
            no = iterador.direita
            iterador._setProximo(None)
            self.lenght -= 1
            
            return no
        
    def _removeMidle(self,indice):
        
        if self._empty():
            return "Lista vazia"
        
        if indice > self.lenght or indice < 0:
            return "Index Error"
        
        if self.lenght == 1:
            return self._removeHead()
        
        else:
            
            contador = 1
            iterador = self.head
            
            while (contador < indice) and (iterador != None):
                
                iterador = iterador.direita
                contador += 1
            
            no = iterador.direita
            iterador._setProximo(no.direita)
            self.lenght -= 1
            
            return no
            
    def _Append(self,no,posicao = None):
        
        if posicao == 0:
            self._insertHead(no)
        elif posicao == None:
            self._insertTail(no)
        else:
            self._insertMidle(no,posicao)
            
            
    def _Del(self,posicao = None):
        
        if posicao == 0:
            self._removeHead()
        elif posicao == None:
            self._removeTail()
        else:
            self._removeMidle(posicao)
```

```python
# Formar lista de adjacencias para representar um grafo

#1 -> 2 - 5
#2 -> 4
#3 -> 4 -6
#4 -> 5
#5 -> 2 
#6 -> 6


# Criar Lista de listas para representar o grafo
listaAdj = [lista(),lista(),lista(),lista(),lista(),lista()]

listaAdj[0]._Append(no(2))
listaAdj[0]._Append(no(5))

listaAdj[1]._Append(no(4))

listaAdj[2]._Append(no(4))
listaAdj[2]._Append(no(6))

listaAdj[3]._Append(no(5))

listaAdj[4]._Append(no(2))

listaAdj[5]._Append(no(6))
```

```python
class Busca:
    
    def __init__(self,listaAdj):
        
        self.listaAdj = listaAdj
        self.visitados = [False,False,False,False,False,False]
        self.pilha = lista()
        self.pai = [None,None,None,None,None,None]
        
    
        
    
    
        
    def _RecursiveBFS(self,inicio):
        
        
        self.visitados[inicio-1] = True
        self.pilha._Append(no(inicio),0)
        no_final = None
        
        while not self.pilha._empty():
            
            atual = self.pilha._removeHead().valor
            self.pilha._Del(0)  
            
            for i in range(self.listaAdj[atual-1]._getLenght()):
                
                if not self.visitados[self.listaAdj[atual-1]._getHead().valor-1]:
                    
                    self.visitados[self.listaAdj[atual-1]._getHead().valor-1] = True
                    self.pai[self.listaAdj[atual-1]._getHead().valor-1] = atual
                    self.pilha._Append(no(self.listaAdj[atual-1]._getHead().valor),0)
                    
                no_final = self.listaAdj[atual-1]._getHead().valor
                
    
        return inicio,no_final
    
    
    
    def _Caminho(self,inicio,fim):
        
        if inicio == fim:
            print(inicio)
        elif self.pai[fim-1] == None:
            print("Nao existe caminho")
        else:
            self._Caminho(inicio,self.pai[fim-1])
            print(fim)
    
    
   
        
        
            
        
        
```
